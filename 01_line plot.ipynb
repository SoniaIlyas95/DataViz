## Adding limits

```python
# step 1 import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# step 2 load data set/or import your own data
flower=sns.load_dataset("iris")
flower
# step 3 draw a line plot[line plot always form between two numeric variable]
sns.lineplot(x="sepal_width", y="sepal_length", data=flower)
plt.title("plot between sepal_length and sepal_width")
plt.xlim(3)
plt.ylim(2)
plt.show()
```

## Change figure size

```python
# step 1 import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# step 2 load data set/or import your own data
flower=sns.load_dataset("iris")
flower
#change figure
plt.figure(figsize=(4,2))
# step 3 draw a line plot[line plot always form between two numeric variable]
sns.lineplot(x="sepal_width", y="sepal_length", data=flower)
plt.title("plot between sepal_length and sepal_width")
plt.show()
```
